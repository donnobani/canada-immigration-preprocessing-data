## Exploring and Pre-Processing: </br> Canadian Immigration Flows


Importing libraries:

```python
import numpy as np
import pandas as pd
```

Reading excel file from URL into a dataframe:

```python
df_canada = pd.read_excel(
    'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
    sheet_name='Canada by Citizenship',
    skiprows=range(20), # non-data rows
    skipfooter=2)
df_canada.head()
```

Retrieving basic information:

```python
df_canada.info(verbose=False)
```

```python
df_canada.columns
```

```python
df_canada.index
```

```python
df_canada.shape
```

#### Removing unnecessary columns:

```python
df_canada.drop(['Type','AREA','REG','DEV','Coverage'],
               axis=1,
              inplace=True)
```

```python
df_canada.head(2)
```

#### Renaming columnns to be more intuitive:

```python
df_canada.rename(columns={'OdName':'Country',
                         'AreaName':'Continent',
                         'RegName':'Region'}, inplace=True)
df_canada.head(2)
```

#### Adding 'Total' column:

```python
# sum of immigration numbers for every year
df_canada['Total'] = df_canada.sum(axis=1, numeric_only=True)
df_canada['Total']
```

#### Checking for null objects in dataset:

```python
df_canada.isnull().sum()
```

No, null objects! Noice!

#### Summary Statistics

```python
df_canada.describe()
```

#### Setting index to Country for easy traversing:

```python
df_canada.set_index('Country',inplace=True)
```

```python
df_canada.head()
```

#### Converting years to strings for congruency:

```python
df_canada.columns = list(map(str, df_canada.columns))
# verifying our conversion worked
# [print (type(x)) for x in df_canada.columns.values]
```

#### Sorting our rows by our 'Total' column:

```python
df_canada.sort_values(by='Total', ascending=False, axis=0, inplace=True)
```
